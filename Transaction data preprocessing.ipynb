```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
directory = 'trans'
new_dir   = 'processed_transactions'
# iterate over files in that directory
for filename in tqdm(os.listdir(directory)):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        tran_n = pd.read_csv(f)
        tran_n = tran_n[~tran_n.MERCHANT.isna()]
        filename = (f.split('\\'))[1].split('.')[0]+'_new'
        tran_n.to_csv(f'{new_dir}/{filename}.csv', mode='a', header=True) 
        #print(filename)
```
